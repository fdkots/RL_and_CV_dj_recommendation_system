```python
# README : Since there are copyright issues, we cant really use a librrary to download the sets. 
# # We did it manually , youtube to mp4 downloader 
# Some very serious fans commented the whole tracklist which we simply copy pasted , and we will map the frames to their corresponding 
# timestamp 
# 
#  

import pandas as pd

tracklist_raw_yusuke_tokyo = """00:45 100 gecs - Dumbest Girl Alive
02:48 Tzusing - Residual Stress
04:38 Skrillex, Nai Barghouti - XENA
08:30 Underworld - Moaner
13:33 The Prodigy - Firestarter (Empirion Mix)
19:50 Locked Club - Kuliki
22:19 Romy, Fred Again.. - Strong
25:26 3Phaz - Drum Track
26:48 For Those I Love - I Have a Love (Overmono Remix)
31:23 The Chemical Brothers - The Darkness That You Fear
35:46 The Chemical Brothers - Free Yourself
39:10 Jamie xx - The Rest Is Noise
43:15 SOPHIE - Whole New World/Pretend World
46:58 Chemotex - Schrade Knives
48:53 Skrillex - With You, Friends (Long Drive)
53:30 Karla Borecky - The Passing of Clouds
56:00 Lorenzo Senni - THINK BIG"""


tracklist_raw_fred_again_london = """
1:10      Fred again.. & Swedish House Mafia - Not Alone (Intro Edit)
1:30      Fred again.. - Kammy (like I do)
6:15      Fred again.. - Danielle (Smile On My Face)
9:52      Fred again.. - Tanya (Maybe Life) 
13:29    Fred again.. - Pull Me Out
18:11    Overmono vs. Lil Baby - BBY vs. Real As It Gets (Fred again.. Bootleg)
21:49    Skrillex & Fred again.. - Stay with me
24:49   Fred again .. - Headie One ft. Skepta - Back to basics
[FLoatinG_pts edit] x Headie One - PTSD
26:35    KH - Feelings
31:34    Fred again.. & Romy - Strong
34:44    Fred again.. & Swedish House Mafia ft. Future - Turn On The Lights Again..
39:09    Fred again.. - Jungle
42:16    Fred again.. & Romy & HAAi - Lights Out w/ Fred again.. & Romy & HAAi - Lights Out (DnB Edit)
47:03    Fred again.. ft. Kodak Black - Kahan (Last Year) w/ Skrillex & Fred again.. - ID *
50:26    Fred Again.. & Skrillex & Four Tet - Baby Again.. *
56:06    Skrillex & Four Tet & Fred again.. ft. Flowdan - Rumble
59:33    Fred again.. & Beverly Craven - Promise me
1:02:20 Fred again.. - Marnie (Wish I Had U) w/ Fred again.. - Faisal (Envelops Me) (Acappella) w/ Fred again.. - ID
1:06:31 Fred again.. - Angie (I've Been Lost) (ID Remix) w/ Fred again.. - ID"""


tracklist_raw_chase_and_status_london = """Boiler Room: London
0:00 Chase & Status & Mozey ft. Sav'o & Horrid1 - On The Block
06:21 Chase & Status ft. Burro Banton - Delete
08:11 T>I - Dye Migration (Serum Remix)
09:28 Chase & Status ft. Popcaan & IRAH - Censor (VIP)
12:07 Mtox & Evoke - Listen Up w/ Watch The Ride & Nia Archives - Mashup The Dance (Kings Of The Rollers Remix)
13:56 CHASE & Status & Hedex ft. Arrdee - Liquor & Cigarettes w/ Hedex - ID
18:40 Chase & Status ft. Cutty Ranks - Retreat2018
20:25 Chopstick Dubplate ft. Ragga Twins & Bunny Lye lye - Give Me A Dubplate
21:30 Dimension - UK (Bou Remix)
23:00 Chase & Status ft. Stefflon Don - Selecta
25:50 TC - Where's My Money (A.M.C & Turno Remix)
27:20 Chase & Status - ID
30:15 Chase & Status ft. Takura - No Problem w/ Rusko & Mozey - Hands In The Air (VIP) w/ Chase & Status ft. Takura - No Problem (VIP)
33:25 ATMOS - Hold You
33:50 Chase & Status ft. Masicka - Weed & Rum (VIP)
38:20 Chase & Status ft. Takura - 2 Ruff
41:00 Chase & Status ft. Irah - Program w/ Adam F - Circles (Pola & Bryson Bootleg)
46:00 UK Apache & Shy FX - Original Nuttah (Chase & Status Remix) /w Digital - Deadline
48:53 Chase & Status ft. Clementine Douglas - Say The Word
53:00 Bou & Chase & Status ft. Trigga & IRAH & Flowdan & Takura - Baddadan (VIP)
"""

tracklist_raw_ketama_london = """ 
00:00 - 05:20 Jump (Rushing mix)- Soul Mass Transit System
05:20 - 09:12 Trip Advisor (BUGS 2025 Warp Mix) - Sydney Charles, BUGS 
09:12 - 13:12 Blessing (KETTAMA remix) - Calvin Harris, KETTAMA
13:12 - 15:40 Skepta Dub -  6 Sense Ned Bennett (Soundcloud)
15:40 - 18:22 Topper top - Sir Spyro (mixed in with something)
18:22 - 19:30 ID - ID 
19:30 - 22:38 De Moma - Raredub (unreleased)
22:38 - 26:37 If U want my heart (Ft. KLP) -  KETTAMA, DJ Heartstring (On Archangel) 
26:37 - 31:57 It gets better (Forever mix) - KETTAMA (On Archangel) 
31:57 - 37:22 ​I need you (Twofaced Sport Remix) - Nikita Warren, TwoFaced (unreleased)
37:22 - 42:05 Man With a Second Face - KETTAMA (On Archangel) 
42:05 - 45:35 Are you ready? - Paque 
45:35 - 50:55 Yosemite - KETTAMA, Interplanetary 
50:55 - 52:07 FODE PRA KARALHO - petrus.wav & MC GW (mixed)
52:07 -56:41 Sort it out - KETTAMA, Clouds (On Archangel) 
"""
tracklist_raw_marlon_hoffstadt_melbourne = """
0:00  Klubbheads - Kickin' Hard (DJ Daddy Trance Extended Remix)
4:48 Andy Farley - Beef Jerky 
10:00 Question Mark - The Birds
14:41 Andy Farley - Out Of Control (Original Mix)
19:37 Patrick Mason - Shred My Panties (Marlon Hoffstadt Edit)
23:29 Darude - Sandstorm (Remix)
26:26 Voodoo & Serano - Blood is pumping
30:07 Vizionn - Run 
33:58 Luciana - I Like That (Marlon Hoffstadt Edit)
38:26 DJ Arne L II vs. DJ Mirko Milano - Cubus (Seven Of Nine)
41:01 Pegassi - Yoyoyo
45:51 Marlon Hoffstadt - Its that Time
50:15  Marlon Hoffstadt - Tell Me Daddy
53:59 Drake - Hotline Bling (Marlon Hoffstadt Remix)
58:26 Malugi, Marlon Hoffstadt - Faster Alone 
1:03:49 Marlon Hoffstadt  - Money On The Dash
1:07:58 Narciss, DJ Heartstring -  Heartbreak Repair System 
1:12:32 Marlon Hoffstadt aka DJ Daddy Trance - Call Me 
1:17:20 Marlon Hoffstadt - I Got You 
1:21:31 Marlon Hoffstadt aka DJ Daddy Trance - Say It Right
"""
tracklist_raw_azyr_teletech_festival = """ 0:07 Smac U - The intro
1:42 Ventur - Bring that beat (Alex Farell Remix)
4:35 Sikoti - You can’t hide
6:35 Sikoti - Bang bang bang
8:39 Byørn - Underground hour
11:46 Veseli - Scream 
13:42 Nyco - Kick the speaker
16:25 Luciid - Green Street (unreleased)
18:39 APHØTIC - Real lies
21:27 Alignment - Close your eyes
24:17 Remon Verhoeve - Patriotic
27:14 Juul Exler - Intrude
29:43 Prax - 1995
31:48 Fiisher - Just a dream
34:50 Alex Farell - Jungle boy 
37:15 Bollman - Rampage (Creeds edit) 
39:09 Azyr - Direct it to the roof (riot codes remix)
42:20 Alex Farell - All night long
45:01 Mahtal - All in this
48:44 Fxck this - Holy Priest
51:29 - Frankyeffe - Save me (RiVid A Bit Harder Edit)
54:50 - Infectious - I need your lovin (95 happy hardcore heavy extended version)
""" 

tracklist_raw_minanokimochi_tokyo = """
0:34 A. G. Cook - Xxoplex
0:50 Lorenzi - Back To You
2:20 DJ Break Da Law - Drosera
4:20 Merely - The Killing Sun (Makinarium Remix)
7:45 TRANCEMAN2000 - Falling van Buuren
10:50 Panteros666 - Meteociel (Dirpix Remix)
11:50 Lyra Valenza - Reality Blizz
13:37 + 16:00 AD†AM - NAK3D (Uphoria Remix)
15:30 Buga - Hands-Up!
17:00 Trillosta - Crystal Sky
18:11 HDMIRROR - ANTHEM
19:00 adaa - feeling a moment
20:00 Sun Angels - A Drink Before The War
22:01 Sun Angels - 777
23:56 Glen - Superfast XI
25:20 Above & Beyond - Morning In Deira
26:46 Sun Angels - Stilla
29:37 Jam & Spoon - Stella’s Cry (Push - Tribute To Markus Remix)
31:51 Nathan Micay - Ecstasy Is On Maple Mountain
32:49 Underworld - Rez
35:30 Marsh - Lailonie
38:00 Emily Glass - MESOGLEA TRANCER WIP☆
39:46 Team Rockit - Team Rockit (Sun Angels Sommarvisa Edit)
42:24 Negative Architecture - Lighter Than Air
43:50 INZO - Earth Magic
""" 

tracklist_raw_yaeji_new_york = """ 
00:15 Etude - Free Tempo
2:00 Freak Dub - Randomer
9:15 PTMC - Officewerks, played with the tokyo drift acapella
10:16 Tokyo Drift 
13:36 Heat - Boddika
20:30 AQXDM - Aegis (+ Teasing Wii in)
24:56 Wii
28:02 Kerosene - Crystal Castles 
36:47 Ifyougotta - bastiengoat
44:54 Raingurl
49:31 Aliyah's Interlude (unreleased😢)
57:35 What's Poppin - Narciss' Champagne Poppin Remix
1:01:06 BOOBOO Don't Touch Me
1:15:00 Running Round - Mercury
1:17:19 The Feeling When You Walk Away - Yves Tumor
1:24:19 I Need You so Much (feat. Roberta Sweed) - Moodymann
"""

tracklist_raw_alisha_DJMAG = """
1. Gene On Earth - Fan Man
2. Chopper - Perfect Day 
3. Coming soon… 
4. Senso x Oncho - ID
5. ID - ID 
6. ID - ID
7. Dan Fresco - Shake The Walls
8. Kolter - No Talking
9. A little bit orange - Sweat
10. Ellia Jaya - Shaking it off 
11. The Trip - Trip Life Xtra
12. Jacob Matthews - Devotion 
13. Sosa - Time Away 
14. L.P Rhythm - Won’t Stop 
15. James View - People of the dance 
16. Jamie Fielding - Aquamarine Edit
17. Ruzie - Lonely 
18. Fish go deep - cure & the cause 
"""


tracklist_raw_estellaBoersma_Berlin = """
57:06 MAXX ROSSI - Get The Funk 
49:01 Disguised - Cortex Me  (mixed nicely with Deas - Focus)
26:41 X CLUB. - RUFF CUTS
16:53 EFESIAN - The Smoke 
43:10 Slam make you move
5:49 Sonic Rain - São Paulo
10:30 Beastie Boys - Intergalactic (AnG Remix)
12:58 X CLUB. - Block Rockin'
20:00 6 SENSE – Funk Contender
28:15 6 SENSE - BASIK 08
28:15 6 Sense x Olly Lishman - HEAVYHITTAS
32:00  Lomax - Gary Beck
34:39  FJAAK  - Redemption
39:00 Schall - Thomas Schumacher Remix
46:00 Cortex Me - distinguised
1:12:20  Mauro Picotto - Baguette
1:15:10 Sped up version of I Am - Alloy Mental
1:26:08 sped-up Insomnia by Faithless (?Tiesto or ?Chris Crime Infinity)
"""

tracklist_raw_peggygou_amsterdam = """ 
00:09 Masters At Work - The Ha Dance [Defected 2010]
05:20 Jessie Ware - Midnight (Peggy Gou's After Midnight Mix) [Universal/Islands Records 2017]
8:30 The Shytowne Project - Everybody's Doing Their Own Thing (ID Remix) [?]
12:08 Hugg & Pepp - Snabeln [Dahlbäck Records 2005]
15:37 Le Gaz Qui Fait Rire (Acid Arab Edit) [Disco Halal 2015]  
19:22 Haruomi Hosono - Laugh-Gas [Epic 1989]
23:14 Philly Slitter - Morning Scent [?]
27:32 Adonis ft. M. C. Kodak - H.O.U.S.E. (Acid Mix) [Black Market Records 1988]
33:19 Alphonse Mouzon with Mouzon's Electric Band - Disco Jungle [Metronome 1979]
35:40 Stopp - I'm Hungry (Caramba Mix) [Discomagic Records 1983]
40:12 Snowboy - Snowboy's House Of Latin [Urban 1989]
45:26 Bad Boy Orchestra - Salsa Time [Smokin’ 1988]
49:57 Third World - Now that We've Found Love (Disco Mix) [Island Records 1978]
54:23 Corporation of One - The Real Life [Smokin’ 1989]
"""

tracklist_raw_yungsingh_melbourne = """ 
0:55 Mura Masa, Erika de Casier - e-motions
3:13 The Streets - Has It Come To This (Pj Bridger Refix)
7:51 Raf-Saperra - N.L.S (Nach Le Soniye)
10:01 Nightcrawlers - Push The Feeling On
12:34 AJD - Gangsta Massive Ft. Karan Aujla, Drake & YG
15:02 Shaggy - It Wasn’t Me (Soul Mass Transit System Remix) 
17:48 Paul Johnson - Get Get Down
21:21 Trotters Independent Traders - Burning
24:27 Double 99 feat. Top Cat - Rip Groove x Heer (ARJXN Dub)
27:14 Rebound X - Rhythm N Gash (Pj Bridger Refix)
28:21 Silver - Sisters Dub
29:30 Smokey Bubblin’ B - Meet Her At The Garage Tent
31:40 T2 feat. Jodie Aysha - Heartbroken
35:08 Sk*n On Sk*n - B*rn D*m Br*dg*s [this line triggers YouTube's spam filter for some reason]
37:40 Vengaboys - Boom, Boom, Boom, Boom!! (ID edit)
39:20 Sash! - Encore Une Fois (Future Breeze Edit)
42:15 DJ Daddy Trance (Marlon Hoffstadt) - Day n Night
45:29 CZETAN - Ik Banere Mashup (Sammy Virji - Blue Roll x Tigerstyle ft. Ms Rajni - Ik Banere)
48:00 Y U QT - Y’all Ready For Dis
50:06 MC D (aka D Double E) & Watch the Ride - Original Format
52:34 M-Beat & General Levy - Incredible
54:22 JK - Panga Peh Giyah x Minor Forms - Versatile Sound
57:02 piri & tommy - on and on
59:26 Vampire Weekend - A-Punk
""" 

tracklist_raw_felixdahousecat_circus = """ 
00:00  Gino Soccio - Dancer ( ??? Remix)
05:50   Felix Da Housecat - Funky Dunk
09:10  Marlon D - Jesus Creates Sound (Main Mix)
13:10  Kristin Velvet - Light Up The Night 
17:20  Hamilton Bohannon - Let's Start The Dance
20:20  Talking Heads - Burning Down The Vcid House (Lvte Nite Tuff Guy) (Pitched Up from 104 to 125)
24:00  Moodymann - Shades of Jae (??? Remix)
27:45  LF System - Be My Babe
31:00  Discotronic - Around7
33:41  Loleatta Holloway - Catch Me On The Rebound (Walter Gibbons 12'' Mix)
36:00  Chuck Robert's  My House Vocal Mixed with 8Bar Loop From the Beggining of the  Next Track
38:10  Gene Farris & Cajmere - Sauer's (Original Mix)
42:00  Tiga - IAMWHATIAM
45:35  Yazoo - Don't Go
47:45  Henrik Schwarz - Take Words In Return (Return Version)
53:45  James Brown - Bodyheat (Part II)
56:30  Danny Ocean - Hoe¦Ê Verwagtinge (Original Mix)
57:50  Felix da Housecat - Frankie Meets Pauline
01:01:28  Roni Griffith - Spys (??? Edit)
01:04:17  Floorplan - Sun In The Sky
01:06:38  ID - ID
01:13:00  Theo Parrish - Lake Shore Drive Mixed With Vocals From First Choice - Let No Man Put Asunder
01:17:47  Funky Junction & KC Flightt - Voices (Lost & Found Original Dub)
01:21:30  Freemasons - Uninvited 2014 (feat. Bailey Tzuke) (??? Remix)
01:24:30  Robert Owens - I'll Be Your Friend (Original Def Mix)
01:28:35  ID - ID
01:30:58  Hot Music - Soho (Pitched Up From 112 to 125)
01:34:30  Jamie 3:26 & Cratebug - Hit It n Quit It
01:37:40  ID - ID
01:40:17  Dangerous Dan & Nicky Night Time & Rhonda INTL. ft. Julian Mitchell - A History (Original Mix)
01:44:10  Gichy Dan's Beachwood - On A Day Like Today (Todd Terje "Friendly Children" Edit)
01:47:30  Jerry C King - I Want You (Jerry C. King's Marvin Rework)
01:52:30  3 Winans Brothers - Dance (Louie Vega Dance Ritual Mix) [feat. The Clark Sisters]
""" 

import re
import pandas as pd

# ── Helpers ──────────────────────────────────────────────────────────────────

def _to_seconds(time_str: str) -> int:
    """M:SS / MM:SS / H:MM:SS → seconds."""
    parts = list(map(int, time_str.strip().split(":")))
    if len(parts) == 2:
        return parts[0] * 60 + parts[1]
    return parts[0] * 3600 + parts[1] * 60 + parts[2]

# Matches:  <start>  [- <end>]  <song text>
#   · start / end accept M:SS, MM:SS, H:MM:SS
#   · end block is optional  (Ketama has it, others don't)
_TIME_RE = re.compile(
    r'^\s*(\d{1,2}:\d{2}(?::\d{2})?)'          # start timestamp
    r'(?:\s*-\s*(\d{1,2}:\d{2}(?::\d{2})?))?'  # optional end timestamp
    r'\s+(\S.*\S|\S)'                           # song (non-empty)
)

def parse_tracklist(raw: str, dj: str, location: str) -> pd.DataFrame:
    rows = []
    for line in raw.splitlines():
        m = _TIME_RE.match(line)
        if not m:
            continue  # skip headers, blank lines, bracket-only lines
        start_str, end_str, song = m.group(1), m.group(2), m.group(3).strip()
        rows.append({
            "dj":              dj,
            "location":        location,
            "timestamp_start": _to_seconds(start_str),
            "_end_explicit":   _to_seconds(end_str) if end_str else None,
            "song":            song,
        })

    df = pd.DataFrame(rows)
    # timestamp_end: explicit (Ketama) > inferred from next song > inf (last track)
    df["timestamp_end"] = (
        df["_end_explicit"]
        .combine_first(df["timestamp_start"].shift(-1))
        .fillna(float("inf"))
    )
    df = df.sort_values("timestamp_start").reset_index(drop=True) #make sure its sorted by time 
    return df.drop(columns=["_end_explicit"])

# ── Parse all sets & concatenate ─────────────────────────────────────────────

sets_metadata = [
    (tracklist_raw_yusuke_tokyo,               "Yusuke Yukimatsu", "Tokyo"),
    (tracklist_raw_fred_again_london,          "Fred Again..",     "London"),
    (tracklist_raw_chase_and_status_london,    "Chase & Status",   "London"),
    (tracklist_raw_ketama_london,              "KETTAMA",          "London"),
    (tracklist_raw_marlon_hoffstadt_melbourne, "Marlon Hoffstadt", "Melbourne"),
    (tracklist_raw_azyr_teletech_festival,     "AZYR"      ,    "Teletech"), 
    (tracklist_raw_yaeji_new_york ,            "Yaeji" ,        "New_York"),
    (tracklist_raw_minanokimochi_tokyo ,       "MInanokimochi" ,  "Tokyo"),
    #(tracklist_raw_alisha_DJMAG ,              "Alisha"      ,     "DJMAG"),
    (tracklist_raw_estellaBoersma_Berlin      ,"EsttelaBoersma" ,  "Berlin"),
    (tracklist_raw_peggygou_amsterdam,         "PeggyGou"      ,   "Amsterdam"),
    (tracklist_raw_yungsingh_melbourne       ,  "YungSingh"     ,  "Melbourne"),
    (tracklist_raw_felixdahousecat_circus      ,  "Felixdahousecat"     ,  "Circus")
]

all_songs_df = pd.concat(
    [parse_tracklist(raw, dj, loc) for raw, dj, loc in sets_metadata ],
    ignore_index=True
)

print(all_songs_df[["dj", "location", "timestamp_start", "timestamp_end", "song"]])

```

```python
# README 
# Attempt 1: Use the already existing MP4s, convert them to mp3, SLICE based on timestamps we have 
# and use essentia library  https://essentia.upf.edu
# for the first part, use ffmpeg 
import subprocess
import numpy as np
import glob, os

VIDEO_DIR = "/Users/fokia/CentraleSupelec-CV-and-RL-project/dj_sets_videos"
AUDIO_DIR = "/Users/fokia/CentraleSupelec-CV-and-RL-project/audio"
os.makedirs(AUDIO_DIR, exist_ok=True)

# ── Step 1: Extract audio from each MP4 ──────────────────────────────────────

def extract_audio(video_path: str, out_path: str):
    subprocess.run([
        "ffmpeg", "-y", "-i", video_path,
        "-q:a", "0", "-map", "a", out_path
    ], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

video_to_audio = {}  # video_name → audio_path
for video_path in sorted(glob.glob(f"{VIDEO_DIR}/*.mp4")):
    video_name = os.path.splitext(os.path.basename(video_path))[0]
    audio_path = os.path.join(AUDIO_DIR, f"{video_name}.mp3")
    print(f"Extracting audio: {video_name}")
    extract_audio(video_path, audio_path)
    video_to_audio[video_name] = audio_path
```

```python
# README : Use essentia 
import essentia.standard as es
import numpy as np

def analyze_segment_essentia(audio_path: str, t_start: float, t_end: float) -> dict:
    loader = es.MonoLoader(filename=audio_path)
    y = loader()
    sr = 44100  # Essentia default

    s = int(t_start * sr)
    e = int(t_end * sr) if t_end != float("inf") else len(y)
    segment = y[s:e]

    if len(segment) < sr:  # skip segments shorter than 1s
        return {}

    bpm, _, _, _, _   = es.RhythmExtractor2013()(segment)
    energy             = float(es.Energy()(segment))
    danceability, _    = es.Danceability()(segment)
    key, scale, key_str = es.KeyExtractor()(segment)
    loudness           = float(es.Loudness()(segment))

    return {
        "bpm": bpm, "energy": energy,
        "danceability": danceability, "key": key,
        "scale": scale, "loudness": loudness,
    }

# Run across all songs (load each audio file once per set)
feature_rows = []
for video_name, group in all_songs_df.groupby("video_name"):
    audio_path = group["audio_path"].iloc[0]
    print(f"Loading: {video_name}")
    
    for _, row in group.iterrows():
        print(f"  ↳ {row['song']}")
        feats = analyze_segment_essentia(audio_path, row["timestamp_start"], row["timestamp_end"])
        feature_rows.append({**row.to_dict(), **feats})

audio_features_df = pd.DataFrame(feature_rows)
print(audio_features_df[["dj", "song", "bpm", "energy", "danceability", "loudness"]].to_string())

```

```python
# Get the rows with missing features
missing_df = audio_features_df.loc[audio_features_df['bpm'].isnull()].copy()
print(f"Re-analyzing {len(missing_df)} missing records...")

FEATURE_COLS = ["bpm", "energy", "danceability", "key", "scale", "loudness"]

def analyze_segment_essentia_padded(audio_path: str, t_start: float, t_end: float) -> dict:
    loader = es.MonoLoader(filename=audio_path)
    y = loader()
    sr = 44100

    s = int(t_start * sr)
    e = int(t_end * sr) if t_end != float("inf") else len(y)
    segment = y[s:e]

    # Pad to at least 1s instead of skipping
    if len(segment) < sr:
        segment = np.pad(segment, (0, sr - len(segment)))

    bpm, _, _, _, _     = es.RhythmExtractor2013()(segment)
    energy               = float(es.Energy()(segment))
    danceability, _      = es.Danceability()(segment)
    key, scale, _        = es.KeyExtractor()(segment)
    loudness             = float(es.Loudness()(segment))

    return {"bpm": bpm, "energy": energy, "danceability": danceability,
            "key": key, "scale": scale, "loudness": loudness}

# Re-run on missing rows
for idx, row in missing_df.iterrows():
    print(f"  ↳ {row['song']} ({row['dj']})")
    feats = analyze_segment_essentia_padded(row["audio_path"], row["timestamp_start"], row["timestamp_end"])
    for col, val in feats.items():
        audio_features_df.at[idx, col] = val

print(f"Remaining nulls: {audio_features_df['bpm'].isnull().sum()}")
```
